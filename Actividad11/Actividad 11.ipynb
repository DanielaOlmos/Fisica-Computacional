```python
#Importamos las bibliotecas
from math import cos
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import ode
```

```python
#Arreglos que contienen cada una de las gráficas
T=[]
grafx = []
grafy = []
        
#Definimos las ecuaciones diferenciales y los parámetros
def f(t, z, p):
    """"
        Ecuaciones diferenciales:
        dx/dt = y
        dy/dt = - δ*y - α*x - β*x**3 + γ*cos(ω*t)
    """
    #Parámetros de la integral
    δ, α, β, γ, ω = p                           
    #Parámetros de integración
    x, y = z                                    
    #Función a integrar
    f = [y, - δ*y - α*x - β*x**3 + γ*cos(ω*t)]  

    return f

#Creamos una instancia para resolver la ecuación diferencial a través del método 'dopri5'
solver = ode(f)
solver.set_integrator('dopri5')

#Especificamos los valores de γ
γi = [0.20,0.28,0.29,0.37,0.50,0.65]

#Mediante este arreglo graficamos cada valor de γ
for i in range(0,len(γi)):

    #Especificamos los parámetros de la ecuación a nuestra función
    γ = γi[i]
    α = -1.0        
    δ = 0.3
    ω = 1.2
    β = 1.0
    val = δ, α, β, γ, ω                      
    solver.set_f_params(val)

    # Fijamos el valor inical z(0) = z0
    t0 = 0.0
    z0 = [1.0, 0]
    solver.set_initial_value(z0, t0)

    #Creamos un arreglo de valores de tiempo t para evaluar la solución; creamos un arreglo que 
    #almacene los valores obtenidos
    t1 = 80
    n = 600
    t = np.linspace(t0, t1, n)
    sol = np.empty((n, 2))

    #Asignamos las condiciones iniciales
    sol[0] = z0

    #Llamamos repetidamente al integrador para que avance en la solución un tiempo 'k'
    k = 1

    while solver.successful() and solver.t < t1:
        solver.integrate(t[k])
        sol[k] = solver.y
        k += 1

    #Guardamos posición y velocidad    
    x = sol[:,0]
    y = sol[:,1]
    
    #Calculamos los valores de T
    Ti=[]
    for i in range (0,len(t)):
        T0= (t[i]) * ω / (2* np.pi)
        Ti.append(T0)

         
    #Guardamos los valores calculados para cada gráfica
    grafx.append(x)
    grafy.append(y)
    T.append(Ti)
```

```python
plt.plot(T[0],grafx[0])
plt.xlabel("t/T")
plt.ylabel("x")
plt.title("Posición contra tiempo (γ=0.20)")
plt.grid(True)
plt.savefig('P1',plt=2000)
```

```python
plt.plot(grafx[0], grafy[0], marker=".")
plt.xlabel('x')
plt.ylabel('ẋ')
plt.title("Retrato de fase (γ=0.20)")
plt.grid(True)
plt.savefig('R1',plt=2000)
```
